# 3.Modeling

## Trainer
提供一個可以重複使用的trainer，在每次不同的特徵有所改變的時候近需要微量的code就可以驗證不同特徵組之間的效果。

```python
import sklearn
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor

BASE_MODEL = RandomForestRegressor(random_state=2, n_estimators=250, min_samples_split=20) 

def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std()) 
            
def train_and_evaluate_model(x, y):
    model = sklearn.base.clone(BASE_MODEL)
    scores = cross_val_score(model, x, y, scoring="neg_mean_squared_error", cv=10)
    train_rmse_scores = np.sqrt(-scores)
    display_scores(train_rmse_scores)  

def model_report(FEATURES, TARGET='SalePrice'):
    global train
    x = train[FEATURES]
    y = train[TARGET]
    train_and_evaluate_model(x, y)

def train_and_submit(features, target, submission_file_path):
    global train
    global test
    model = sklearn.base.clone(BASE_MODEL)
    model.fit(train[features], train[target])
    submit_df = pd.DataFrame(index=test.Id)
    submit_df['SalePrice'] = model.predict(test[features])
    submit_df.to_csv(submission_file_path)
```

## 3-1.Baseline model
影響房價的因素主要為環境以及房屋的屬性，基線模型以坪數以及房屋品質來當作特徵，從這裡出發再逐漸加入其他特徵，如果之後加入的特徵表現更差的話表示新增的特徵中有噪聲或與房價的關聯不大。

```python
BASELINE_FEAT = ['OverallQual', 'GrLivArea']
```

```python
plt.figure(figsize=(21, 7))
plt.scatter(train['OverallQual'], train['SalePrice'], alpha=0.2);
plt.show()
```

```python
outlier_indices = train[(train.OverallQual == 10) & (train.SalePrice < 200000)].index
train = train[~train.index.isin(outlier_indices)]
```

```python
plt.figure(figsize=(21, 7))
plt.scatter(train['GrLivArea'], train['SalePrice'], alpha=0.2);
plt.show()
```

```python
outlier_indices = train[(train.GrLivArea > 4000)].index
train = train[~train.index.isin(outlier_indices)]
```

### 模型測試報告

```python
model_report(FEATURES=BASELINE_FEAT)
```

```python
train_and_submit(features=BASELINE_FEAT, target='SalePrice', submission_file_path='submissions/baseline_model.csv')
```

## 添加屋齡為特徵
直觀的來說屋齡對我們也是個重要的衡量屋價的標準，以下對此特徵做EDA

```python
train['HouseAge'] = train['YrSold'] - train['YearBuilt']
test['HouseAge'] = test['YrSold'] - test['YearBuilt']
```

```python
Feat = ['OverallQual', 'GrLivArea', 'HouseAge']
model_report(FEATURES=Feat)
train_and_submit(features=Feat, target='SalePrice', submission_file_path='submissions/model_v1.csv')
```
